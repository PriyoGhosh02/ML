```python
import pandas as pd
from sklearn import linear_model
from word2number import w2n
import numpy as np
```

```python
# !pip install word2number
```

```python
df=pd.read_csv("/kaggle/input/salaryonexp/hiring.csv")
# df
```

```python
df["experience"]=df["experience"].fillna("zero")
df["experience"]=df["experience"].apply(w2n.word_to_num)
# df
```

```python
df["test_score(out of 10)"]=df["test_score(out of 10)"].fillna(df["test_score(out of 10)"].mean())
df
```

```python
model=linear_model.LinearRegression()
model.fit(df[["experience","test_score(out of 10)","interview_score(out of 10)"]],df["salary($)"])
```

```python
model.coef_
```

```python
model.intercept_
```

```python
model.predict(pd.DataFrame([[2, 9.00, 6]], columns=['experience', 'test_score(out of 10)', 'interview_score(out of 10)']))
```
